train_test_split:
We split the whole dataset into trainset and testset which contains 75% train and 25% test.

We can include this train set into classifiers to train our model and the test set is useful for predicting the performance of the model by different classifiers.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import RandomizedSearchCV
```

```python
x_train, x_test,y_train, y_test=train_test_split(x,y,test_size=0.25,random_state=40)
#test_size parameter is used to specify the proportion of the dataset that should be included in the testing set.
#if you set test_size to 0.25, it means that 25% of the data will be used for testing, and the remaining 
#75% will be used for training.
# x is the input data and y is the target variable.
#random_state parameter is also set to 40, which ensures that the same random 
#split will be generated each time the function is run 
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Decision tree

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import cross_val_score, GridSearchCV 
from sklearn.metrics import classification_report 
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import f1_score
```

```python
tree_model = DecisionTreeClassifier(max_depth=5,criterion='entropy')
# max_depth parameter is used to prevent overfitting by limiting the maximum depth of the decision tree.
#For  quality of the split we use criterion entropy which measures the degree of randomness or disorder in the subset.
# low value indicates homogeneous high value indicates hetrogeneous
cv_scores = cross_val_score(tree_model, x, y, cv=10, scoring='accuracy')
#The cross_val_score function is a utility function which is commonly used for cross-validation. The cv parameter in 
#cross_val_score specifies the number of folds to use for cross-validation.
#The scoring parameter in the cross_val_score function is used to evaluation metric to use when performing cross-validation.
#The scoring parameter is set to 'accuracy', which means that the accuracy score will be used to evaluate the performance of the
#model.
m=tree_model.fit(x, y)
# The fit() function takes as input the training data and the target variable and learns the relationship between them, which 
#can then be used to make predictions on new, unseen data.
prediction=m.predict(x_test)
cm= confusion_matrix(y_test,prediction)
sns.heatmap(cm, annot=True,cmap='winter',linewidths=0.3, linecolor='black',annot_kws={"size": 20})
print(classification_report(y_test, prediction))
```

```python
TP=cm[0][0]
TN=cm[1][1]
FN=cm[1][0]
FP=cm[0][1]
```

```python
print('Testing Accuracy for Decision Tree:',(TP+TN)/(TP+TN+FN+FP))
print('Testing Sensitivity for Decision Tree:',(TP/(TP+FN)))
print('Testing Specificity for Decision Tree:',(TN/(TN+FP)))
print('Testing Precision for Decision Tree:',(TP/(TP+FP)))
```

From the above report, we get the accuracy of the Decision Tree classifier is about 92%.



KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
list_1=[]
for i in range(1,21):
    knn=KNeighborsClassifier(n_neighbors=i)  
    knn.fit(x_train,y_train)
    preds=knn.predict(x_test)
    scores=accuracy_score(y_test,preds)
    list_1.append(scores)
```
